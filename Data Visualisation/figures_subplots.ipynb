```python
# Figure
fig = plt.figure()

# Added subplot inside the Figure
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# Histogram plot
# color
# alpha is transparency 
ax1.hist(np.random.standard_normal(100), bins=20, color="blue", alpha=0.2);

# Scatter plot
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30), color='black', alpha=0.3);

# Plot
ax3.plot(np.random.standard_normal(50).cumsum(), color="black", linestyle="dashed")
```

## Removing the spaces between the subplots

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
 for j in range(2):
     axes[i, j].hist(np.random.standard_normal(500), bins=50, color="black", alpha=0.5)
fig.subplots_adjust(wspace=0, hspace=0)

```
